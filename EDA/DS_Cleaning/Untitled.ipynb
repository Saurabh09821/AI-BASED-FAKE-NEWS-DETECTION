```python
import pandas as pd

df = pd.read_excel("F:/PG-DBDA-2025/Project_Upload/Data_using/bharatfakenewskosh.xlsx")
```

```python
import re
import pandas as pd

def clean_publish_date(text):
    if pd.isna(text):
        return pd.NaT
    text = str(text)
    text = re.sub(r'(\d+)(st|nd|rd|th)', r'\1', text)
    try:
        return pd.to_datetime(text, dayfirst=True, errors='coerce')
    except:
        return pd.NaT

df["publish_date_parsed"] = df["Publish_Date"].apply(clean_publish_date)
```

```python
df["Label"] = pd.to_numeric(df["Label"], errors="coerce")
```

```python
df = df.drop_duplicates()
```

```python
df["Eng_Trans_Statement"] = df["Eng_Trans_Statement"].fillna("")
df["Eng_Trans_News_Body"] = df["Eng_Trans_News_Body"].fillna("")
```

```python
df["statement_word_count"] = df["Eng_Trans_Statement"].apply(lambda x: len(str(x).split()))
df["body_word_count"] = df["Eng_Trans_News_Body"].apply(lambda x: len(str(x).split()))
```

```python
df.to_csv("cleaned_full_dataset.csv", index=False)
```
